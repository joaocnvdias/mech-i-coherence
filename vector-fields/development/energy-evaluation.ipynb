```python
import json
import matplotlib.pyplot as plt
import statistics
```

```python
def read_json_list(filepath):
    with open(filepath, 'r') as f:
        return json.load(f)  

file1 = 'checkpoints/energy_viktor.json'
file2 = 'checkpoints/energy_prague.json'
file3 = 'checkpoints/energy_sciencefic.json'
file4 = 'checkpoints/energy_inc1.json'
file5 = 'checkpoints/energy_inc2.json'
file6 = 'checkpoints/energy_inc3.json'
file7 = 'checkpoints/energy_inc1_newgen.json'
file8 = 'checkpoints/energy_inc2_newgen.json'


viktor = read_json_list(file1)
prague = read_json_list(file2)
science_fic = read_json_list(file3)
inc1 = read_json_list(file4)
inc2 = read_json_list(file5)
inc3 = read_json_list(file6)
inc1_newgen = read_json_list(file7)
inc2_newgen = read_json_list(file8)
```

```python
#original prompts
plt.figure(figsize=(12, 8))  


plt.plot(viktor, label='Viktor (man-made)')
plt.plot(prague, label='Prague (LLM-prompt)')
plt.plot(science_fic, label='Science Fiction (LLM-prompt)')

plt.xlabel('Story Index')
plt.ylabel('Total Prompt Energy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#llm coherent
plt.figure(figsize=(12, 8)) 

plt.plot(prague, label='Prague (LLM-prompt)')
plt.plot(science_fic, label='Science Fiction (LLM-prompt)')

plt.xlabel('Story Index')
plt.ylabel('Total Prompt Energy')
plt.legend()
plt.grid(True)
plt.ylim(60.5,64)
plt.show()
```

```python
#LLM incoherent
plt.figure(figsize=(12, 8))  

plt.plot(inc1, label='Incoherent 1')
plt.plot(inc2, label='Incoherent 2')
plt.plot(inc3, label='Incoherent 3')
plt.xlabel('Story Index')
plt.ylabel('Total Prompt Energy')
plt.legend()
plt.grid(True)
plt.ylim(60.5, 64)
plt.show()

```

```python
var_viktor = statistics.variance(viktor)
var_prague = statistics.variance(prague)
var_science_fic = statistics.variance(science_fic)
var_inc1 = statistics.variance(inc1)
var_inc2 = statistics.variance(inc2)
var_inc3 = statistics.variance(inc3)
var_inc1new = statistics.variance(inc1_newgen)
```

```python
variances = {
    'viktor': statistics.variance(viktor),
    'prague': statistics.variance(prague),
    'science_fic': statistics.variance(science_fic),
    'inc1': statistics.variance(inc1),
    'inc2': statistics.variance(inc2),
    'inc3': statistics.variance(inc3),
    'inc1_newgen': statistics.variance(inc1_newgen),
    'inc_2_newgen':statistics.variance(inc2_newgen)
}
```

```python
#all stories
labels = list(variances.keys())
values = list(variances.values())

plt.figure(figsize=(10, 6))
plt.bar(labels, values, color='steelblue')
plt.ylabel('Variance')
plt.title('Variance of Different Generations')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
